```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/submissions.csv")
```

```python
df.columns
```

```python
fig, ax = plt.subplots()

ax.plot(df["score"], "o-")

ax.set_xticks(range(len(df)))

ax.set_xlabel("Submission")
ax.set_ylabel("Score (Accuracy)")
ax.set_title("Model Performance Across Submissions: Titanic Survival Prediction", pad=10)

ax.grid(alpha=0.3)
```
